Potential forms are used when expressing the performance of an action or

## 可 能 形 (Potential Form)

### 1. Explanation
The potential form in Japanese is used to express the **ability, possibility, or potential** to perform an action or achieve a certain state. It indicates that something is feasible or achievable under given circumstances.

### 2. Conjugation/Transformation Rules
* **Verb Conjugation:**
  - **Group 1 verbs:** Add "られる" (rareru) to the plain form.
    - Example: 食べる (taberu) → 食べられる (taberareru)
  - **Group 2 verbs:** Add "られる" (rareru) to the plain form, but change the final "る" to "れる".
    - Example: 書く (kaku) → 書ける (kakeru)
  - **Group 3 verbs:** Add "られる" (rareru) to the plain form, but change the final "る" to "れる" and the vowel in the penultimate syllable to its corresponding "u" sound.
    - Example: する (suru) → できる (dekiru)

### 3. Examples and Explanation

### 4. Important Points
* **Ability:** Expresses the ability to do something.
* **Possibility:** Indicates that something is possible or feasible.
* **Permission:** Can imply permission or allowance.
* **Negative Form:** The negative form (e.g., "眠れない" (nemurenai)) indicates inability.
* **Passive Voice:** The potential form can sometimes be used to express a passive voice meaning.
