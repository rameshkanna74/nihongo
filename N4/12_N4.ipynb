1. 命 令 形
   Imperative form
   The imperative form is a verb form often used when strongly ordering someone.

   It is often used when accidents, disasters, and emergencies occur, but it is

### 1. **Explanation:**

The imperative form (命令形, **meireikei**) is used in Japanese to issue strong commands or orders. This form can be seen in various contexts, such as during emergencies, while encouraging someone, or when giving instructions. It is crucial to note that using the imperative form can come off as rude or disrespectful if directed toward superiors or those with whom you have a formal relationship. Therefore, it’s essential to be mindful of the social context when employing this form.

### 2. **Conjugation/Transformation Rules:**

- **Group 1 Verbs (u-verbs)**: Change the last kana of the dictionary form into the corresponding kana from the **え** column of the same row.

- **Group 2 Verbs (ru-verbs)**: Change the dictionary form ending **る** into **ろ**.
  - Example: **食べる (taberu)** → **食べろ (tabero)**
- **Group 3 Verbs (irregular verbs)**: Memorize the imperative forms.
  - Example: **する (suru)** → **しろ (shiro)**, **来る (kuru)** → **来い (koi)**
