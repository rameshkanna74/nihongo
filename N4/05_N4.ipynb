# Grammar 1


1. 意 向 形
Volitional form
The volitional form of verbs is used to express the subject's intention and will,
such as when asking someone a question.
It is also frequently used when emphasizing an opinion.

    1.明日 私 は 休 みなので、 一 緒 に 映画を 見に 行こうか。
    Tomorrow is my day off, shall we go watch a movie together?

    2.A: 来 週 引っ越しします。
    I will move next week.

    B: 手伝おうか。
    Do you need my help?

    3. 山田さんは まだ 来ていないね。 彼 に 電話しよう。
    Mr. Yamada hasn't come yet. I will call him.

### 1. **Explanation:**

The **Volitional Form** (意向形, ikoukei) of Japanese verbs is used to express the speaker's intention or will to perform an action. This form is particularly useful in making suggestions, offering assistance, or stating plans. It can also indicate the speaker's desire or resolution. By using the volitional form, speakers can ask others to join them in an activity or indicate what they plan to do.

### 2. **Conjugation/Transformation Rules:**

The volitional form is created differently depending on the verb group:

# Grammar 3

This expresses an assumed condition. The “た” of “たら” is the past tense of the plain form.

### 1. **Explanation:**

# Grammar 6


6.V dictionary form 予定です。/ N の予定です。

I plan to do .... / a plan of ...

This conveys a fixed schedule that is going to happen regardless of the speaker's intention.

“~つもり” conveys a personal intention, while “予定です” conveys a decision made by other people.

1. 3 月 10 日に アメリカへ 出 張 する 予定です。
   I plan to go to America on business on March 10th.

2. 展 覧 会 に 参加する 予定です。
   We plan to attend an exhibition.

3. 明日は 休 みの 予定です。
   Tomorrow is scheduled to be a day off.


### 1. **Explanation:**

The construction **V dictionary form 予定です** or **N の予定です** is used to express a scheduled plan or arrangement. It indicates that something is set to happen, often with an implication of formality or commitment. Unlike **つもり**, which expresses a personal intention, **予定** implies a decision that is typically made by oneself or others, suggesting a more fixed or formal arrangement.

### 2. **Conjugation/Transformation Rules:**

To form this structure:

- For verbs, use the dictionary form followed by **予定です**.
- For nouns, use the noun followed by **の予定です**.

**Examples:**

- **行く** (iku) → 行く予定です (iku yotei desu) → I plan to go.

# Grammar 7


7.助詞 + も
Particle + も

“も” can be used with the particle “に”, “で”, “から”, “まで”, “と” and “へ”.

In this occasion, “も” is used to list things or to emphasize something unexpected.

When “も” is used with an interrogative, it expresses total affirmation or total denial.

1. 彼 は 毎 日 野 球 を 練 習 しています。 雨 の日にも 練 習 しています。
   He practices baseball every day. Even on rainy days, he practices.

2. 忙 しい 社 長 が、飲み 会 にも 参加しました。
   Our busy company president still attends the drinking party.

3. 足 を 怪我したので、どこへも 行けません。
   Because I injured my foot, I can't go anywhere.


### 1. **Explanation:**

The particle **も** is used in Japanese to indicate "also" or "too." It can be attached to various particles, such as **に** (ni), **で** (de), **から** (kara), **まで** (made), **と** (to), and **へ** (e). When used in this way, **も** can express inclusion, unexpectedness, or total affirmation/denial when combined with interrogatives.

### 2. **Conjugation/Transformation Rules:**

- **も** is added directly after the particle it modifies. For example:
  - **に** + **も** → **にも**
  - **で** + **も** → **でも**
  - **から** + **も** → **からも**
  - **まで** + **も** → **までも**
  - **と** + **も** → **とも**
  - **へ** + **も** → **へも**

# Grammar 8

The structure **V[ます-form]やすいです** and **V[ます-form]にくいです** is used to express how easy or difficult it is to perform an action denoted by the verb.
