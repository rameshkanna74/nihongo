### Explanation and Important Points:

1. **Usage**:

   - **To Explain Reasons**: Use “...んです” to explain why something happened or provide context.

     - **Example**: **「遅れたんです。」** (I was late because...).

   - **To Provide Context**: It gives additional information or context about a situation or a condition.
     - **Example**: **「お腹が空いたんです。」** (I’m hungry; it’s because I haven’t eaten).

2. **Forms**:

   - **Verbs**: Use the plain form of the verb + “んです” to explain why you did or didn’t do something.

     - **Example**: **「勉強しないんです。」** (I’m not studying because...).

   - **い-Adjectives**: Drop the final **い** and add **そうです**.

     - **Example**: **「このケーキは甘いんです。」** (This cake is sweet).

   - **な-Adjectives**: Add **な** before **んです**.

     - **Example**: **「彼は静かなんです。」** (He is quiet).

   - **Nouns**: Add **なんです** after the noun.
     - **Example**: **「私の趣味は読書なんです。」** (My hobby is reading).

3. **Contrast with Other Patterns**:

   - Unlike **...です** which states facts or information, **...んです** is more explanatory and provides reasons or background.
   - **...のです** is more formal and often used in written language.

4. **Questions**:

   - To ask why something is the case or to request an explanation, use **...んですか**.
     - **Example**: **「どうして遅れたんですか？」** (Why were you late?).

5. **Contextual Clarity**:
   - This grammar pattern is particularly useful in conversational Japanese for clarifying reasons or motivations behind actions or states.

### Practice Examples:

1. **To Explain Actions**:

   - **「遅刻したんです。電車が遅れたんです。」** (I was late because the train was delayed.)

2. **To Provide Context**:

   - **「この映画は面白いんです。」** (This movie is interesting because...).

3. **To Clarify Personal State**:
   - **「今日は疲れているんです。」** (I’m tired today; that’s why I’m not coming to the party.)
