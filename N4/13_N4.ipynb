  - **ございます (gozaimasu)**: A more formal version of "to be."
  - **ましょう (mashou)**: Used to suggest or invite action (let's).
  - **でしょう (deshou)**: Indicates probability or conjecture (probably).

2. 尊 敬 語
   Respectful expressions
   Respectful expressions show respect to others of a higher rank.

Verbs, nouns, adjectives, and adverbs all have respectful forms.

Especially, respectful verbs, which are classified into three types:

    1. Verbs' respectful form,
    2. Special verbs,
    3. お V [ます-form] に なります

1. 鈴 木 先 生 、この 本 使 われますか。
   Do you use this book, Mr. Suzuki?

2. おたばこを やめられたんですか。
   Have you given up Tobacco?

3. 昨日の 会議で 社 長 は どんなことを 話 されましたか。
   What did the company president talk about in yesterday's meeting?


### 1. Explanation:

尊敬語 (sonkeigo), or respectful expressions, are used in Japanese to show respect towards individuals of higher status or authority. This form of speech is particularly significant in Japanese culture, where hierarchy and respect play a crucial role in interpersonal communication. By using respectful expressions, speakers convey politeness and deference, which are valued traits in both personal and professional contexts.

Respectful expressions encompass various parts of speech, including verbs, nouns, adjectives, and adverbs. They are especially important for verbs, which can be classified into three types:

1. **Regular respectful forms** (modifications of standard verbs).
2. **Special verbs** that inherently carry a respectful nuance.
3. **お V [ます-form] に なります**, a structure that elevates the verb's respectfulness.

### 2. Conjugation/Transformation Rules:

- **Verbs' Respectful Forms**:

  - Many common verbs have a specific respectful form. For example:
    - 行く (iku) → いらっしゃる (irassharu)
    - 来る (kuru) → いらっしゃる (irassharu)
    - 見る (miru) → ご覧になる (goran ni naru)

- **Special Verbs**:

  - Certain verbs are already respectful and should be used in their standard form:
    - いる (iru) → いらっしゃる (irassharu)
    - する (suru) → なさる (nasaru)

- **お V [ます-form] に なります**:
  - To elevate a verb to a respectful form:
    - For example, 食べます (tabemasu) → お食べになります (o-tabe ni narimasu).

# Grammar 4


4.V [て-form] いただきます。

a polite way to say “て もらう”
a polite way to say “て くれる”

The plain forms are “V [て-form] いただく” and “V [て-form] くださる”.

This structure expresses that an action benefits the receiver.

Setting A as the receiver, and B as the giver, the following is the correct structural pattern:

Here’s a detailed explanation of the grammar point **V [て-form] いただきます / V [て-form] くださいます**.

### 1. **Explanation:**

The expressions **V [て-form] いただきます** and **V [て-form] くださいます** are polite ways to express that an action benefits the speaker (in the case of いただきます) or the listener (in the case of くださいます). This structure highlights respect and humility in interpersonal interactions.

- **いただく** is the humble form of **もらう** (to receive).
- **くださる** is the respectful form of **くれる** (to give).

This grammar is especially significant in business contexts, where showing respect and humility is crucial.

### 2. **Conjugation/Transformation Rules:**

1. Convert the verb into its て-form.

- **見る (to see)** → 見て (mite)

# Grammar 6


6. ...場合は、...

If...,...

It means a situation happens hypothetically or generally.

“場合” is a noun, so the word that precedes it, is modified.

V dictionary form 場合は、...
V た-form 場合は、...
い-adj 場合は、...
な-adj な 場合は、...
N の 場合は、...


1.明日の 飲み 会 に 行く場合は、 私 に 連 絡 してください。
If you decide to go to the drinking party tomorrow, please contact me.

2. 雨 が 降った場合は、イベントは 中 止 します。
If it rains, the event will be cancelled.

3.明日 忙 しい場合は、山田さんに 手伝って もらいます。
If tomorrow is busy, I will ask Mr. Yamada to help me.

4. 説 明 書 が 必 要 な 場合は、この引き出しの 中 を 探 してください。
If you need the manual, please look for it in this drawer.

5.子供の 場合は、チケットは 半 額 です。
In the case of children, tickets are half price.

### 1. Explanation:

The phrase **...場合は、...** (baai wa) is used to express hypothetical or general situations in Japanese. It translates to "if...," "in the case of...," or "when..." in English. This structure allows speakers to discuss potential outcomes based on certain conditions. It is commonly used in both spoken and written Japanese, making it a versatile grammatical form.

### 2. Conjugation/Transformation Rules:

To construct sentences using **...場合は、...**, the preceding phrase can be in different forms depending on the context:

- **Verb Dictionary Form**: V dictionary form 場合は (baai wa)
  - Example: 行く場合は (iku baai wa) - If (you) go...
- **Verb Past Tense**: V た-form 場合は (baai wa)

  - Example: 行った場合は (itta baai wa) - If (you) went...

- **い-adjective**: い-adjective + 場合は (baai wa)

  - Example: 暇な場合は (hima na baai wa) - If (you are) free...

- **な-adjective**: な-adjective + 場合は (baai wa)

  - Example: 元気な場合は (genki na baai wa) - If (you are) healthy...

- **Noun**: Noun + の場合は (no baai wa)
  - Example: 子供の場合は (kodomo no baai wa) - In the case of children...
