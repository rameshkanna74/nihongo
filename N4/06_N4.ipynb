# Grammar 1


1.自動詞 と 他動詞
Intransitive verbs and Transitive verbs
Japanese verbs can be separated into two verbs types; intransitive and
transitive. Many of these verb pairs utilize the same Kanji.

1. Intransitive verbs: Verbs where only subject words are attached.

(Example: 開 く、消える)

Objects cannot be added to intransitive verbs, and therefore cannot be used with the particle “を”.
Instead they are used with the particles “に”, “へ”, “と”, and “が”.

    ドアが 開 く,
    学 校 に 行く,
    友 達 と 会う

2. Transitive verbs: Verbs that use both subjects and objects.

(Example:開ける、消す)

Since both objects and subjects are utilized, transitive verbs always use the particle “を”.

ドアを 開ける, ご飯 を 食べる, テレビを 見る.


## 1. **Explanation:**

In Japanese, verbs are categorized into two main types: **intransitive verbs (自動詞, jidōshi)** and **transitive verbs (他動詞, tadōshi)**. Understanding the difference between these two types is crucial for constructing sentences correctly, as they dictate how subjects and objects interact within a sentence.

- **Intransitive verbs** express actions or states that do not require a direct object. They often focus on the subject performing the action by itself, and the particles used are typically “が,” “に,” “へ,” and “と.”
- **Transitive verbs**, on the other hand, indicate actions that require a direct object. They are accompanied by the particle “を” to mark the object of the action.

This distinction is significant because it affects sentence structure, meaning, and the way we express relationships between actions and subjects/objects.

## 2. **Conjugation/Transformation Rules:**

### Conjugation of Intransitive and Transitive Verbs

1. **Intransitive Verbs**:

   - Generally follow the same conjugation rules as regular verbs.
   - Examples:
     - **開く** (あく, aku) – to open (intransitive)
       - ます form: 開きます (あきます, akimasu)
       - Negative form: 開かない (あかない, akanai)

2. **Transitive Verbs**:
   - Conjugation rules are similar, but they require an object.
   - Examples:
     - **開ける** (あける, akeru) – to open (transitive)
       - ます form: 開けます (あけます, akemasu)
       - Negative form: 開けない (あけない, akenai)

### Common Patterns:

- For **Group 1 (U-verbs)**, the final “u” sound changes to “i” in the ます form.
- For **Group 2 (Ru-verbs)**, simply replace the “ru” with “masu.”

# Grammar 5


5.しか
Only

“しか is placed after a noun. Both“しか”and “だけ”are particles that set limitations and have the same meaning. 

However, “しか” is always used with negative forms.

“だけ” is used for both affirmative and negative forms.

Example:

1. 待つしかない。
   I can only wait.

2. この仕事は 山田さんしか 出来ないです。

Here’s a detailed explanation of the Japanese grammar point **しか**:

### 1. **Explanation:**

**しか** is a particle used to express limitations or exclusivity in a negative context. It indicates that something is the only option available, and it is always followed by a negative verb form. This differs from **だけ**, which can be used in both affirmative and negative sentences. The significance of **しか** lies in its ability to emphasize the scarcity or sole availability of an item or action.

### 2. **Conjugation/Transformation Rules:**

- **Structure**: Noun + しか + negative verb form
- **Example**:
  - **待つ** (to wait) becomes **待たない** (not to wait).
  - **この仕事は山田さん** (this work is Mr. Yamada) + **しか** + **できないです** (cannot do) → **この仕事は山田さんしかできないです**.

6.まだ affirmative

Still/as of yet/more/besides

There are three common usages.

1. a status that remains the same as the previous status, meaning “still”.
2. a status that has yet to become an expected state, meaning “still/as of yet”.
3. a status that has some incoming information, meaning “more/besides”.
