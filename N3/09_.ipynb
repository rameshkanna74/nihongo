   It is used to emphasize that something is not true.

### 1. Explanation:

- **V[て-form] なんかない**: Used for verbs in their て-form.

- **N なんかじゃない**: Used for nouns, following the noun with なんかじゃない.

# Grammar 4


4. … 間 /… 間 に
During …

“… 間 ” represents that a state or an action is continuing for a certain period of time.

Example: 休 みの 間 、ずっと実家にいました。

### 1. **Explanation:**

The grammar point **…間 (あいだ) / …間に (あいだに)** indicates a period during which a certain action or state occurs. It is used to emphasize the continuity of an action or situation over a specified duration. This can refer to timeframes in the past, present, or future. It conveys that one event or action is happening while another is ongoing, highlighting the overlap of these activities.

### 2. **Conjugation/Transformation Rules:**

- **Noun + の間**: Used when referring to a specific noun.
  - Example: 休みの間 (during the vacation)
- **Verb [て-form] + いる間**: Used for actions that are currently ongoing.
  - Example: 勉強している間 (while studying)
- **Verb plain form + 間に**: Used to indicate that something happens during a specific timeframe.
  - Example: 食べる間に (while eating)

  - Take the verb stem and add **きれない**.

Here’s a detailed explanation of the grammar point **…ように見える**:

### 1. **Explanation:**

The phrase **…ように見える** is used to express that something appears to be a certain way based on observation or appearance. It reflects the speaker's judgment or perception rather than a definite statement of fact. This construction can be applied to verbs, adjectives, and nouns, allowing for a flexible expression of appearance.

### 2. **Conjugation/Transformation Rules:**

The phrase can be conjugated based on the type of word being used:

- **Verb (plain form) + ように見える**

- **I-adjective (drop い) + ように見える**
  - Example: 高い (takai) → 高いように見える (takai you ni mieru) - looks tall.
- **Na-adjective (add な) + ように見える**
  - Example: 静か (shizuka) → 静かなように見える (shizuka na you ni mieru) - looks quiet.
- **Noun + の + ように見える**
  - Example: 学生 (gakusei) → 学生のように見える (gakusei no you ni mieru) - looks like a student.

Here's a detailed explanation of the Japanese grammar point **…くらい/ぐらい** along with examples and practice exercises.

### 1. **Explanation:**

The expressions **くらい (kurai)** and **ぐらい (gurai)** are used to indicate approximate amounts, degrees, or extents. They can translate to "about," "around," or "so... that" in English, depending on context.

- **ぐらい** is also casual and can be used interchangeably with **くらい**, but is often preferred when referring to time or distance.

### 2. **Conjugation/Transformation Rules:**

- **い-adj くらい/ぐらい**: Adjective in its dictionary form + くらい/ぐらい
