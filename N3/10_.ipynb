1. …っけ
…, is it?

This is used when a speaker wants to re-verify something that they forgot or are unsure of with someone.

V[た-form] っけ

い-adj (…い)かった っけ

“…っけ” is used in conversations with close friends.

# Grammar 2


2.もの/もん

Because …

“…もの” comes at the end of a sentence and is often used to state one's own opinions and reasons.

The casual form is “…もん”.

V plain form (んだ) もの/もん
い-adj (んだ) もの/もん
な-adj だ (なんだ) もの/もん
N だ (なんだ) もの/もん

Why didn't you attend the meeting yesterday?

B:だって、 誰 も知らせてくれなかったんだもの。
Because nobody informed me about it.

2.A:あの子のわがままを 毎 回 許 しちゃだめだよ。
You can't ignore that child's selfish behavior every time.

B:しょうがないよ。ただの子どもなんだもん。
It can't be helped. He's just a kid.

3.A:あれ？ 今日はデートに行くんじゃないの？ どうして 家 にいるの？
Huh? You are going on a date today, aren't you?
Why you are still at home?

I canceled the date.
Because I just had an argument with my boyfriend yesterday.


### 1. **Explanation:**

The grammar point **もの/もん** is used to express reasons or explanations for one’s actions or opinions. It adds a personal touch and is often employed in casual conversations, particularly among friends or family. The usage of **もの** indicates a sense of justification or reasoning for a statement, implying that the speaker is providing a valid excuse or explanation.

### 2. **Conjugation/Transformation Rules:**

- **V plain form (んだ) もの/もん**:
  - Example: 行くんだもん (I’m going because...)
- **い-adj (んだ) もの/もん**:

- **な-adj だ (なんだ) もの/もん**:
  - Example: きれいなんだもん (It's beautiful because...)

3.ものだ
It is used to express common things or a natural condition of something.
There are 4 ways of using “ものだ”.

①It indicates something general, natural or normal, meaning that it
expresses common sense.

V plain form ものだ

い-adj ものだ

な-adj な ものだ

1.日本の 夏 は 暑 いものだ。
Summer in Japan is hot.

2. 人 に 助 けてもらった 時 は、お 礼 を言うものだ。
   It is natural to thank people when they help you.

3. 学 校 で 先 生 に会ったら、 挨 拶 するものだ。
   It is common to greet teachers when you meet them at school.


Sure! Here’s a detailed explanation of the grammar point **ものだ**.

### 1. Explanation:

**ものだ** is used to express common truths, natural states, or accepted norms. It reflects generalizations or things that are considered to be natural, usual, or logical in a given context. This expression can also convey an emotional or nostalgic nuance, emphasizing that something is inherent or typical.

### 2. Conjugation/Transformation Rules:

**ものだ** can be attached to different forms of verbs and adjectives:

- **V plain form + ものだ** (for verbs)
- **い-adj (dictionary form) + ものだ** (for adjectives)
- **な-adj (な form) + ものだ** (for na-adjectives)
- **Noun + だ + ものだ** (for nouns)

### 1. **Explanation:**

The grammar point **ものだ** can be used to express a strong desire or wish regarding a situation or outcome. It emphasizes the speaker's longing or hope for something to happen. This construction is commonly used in conjunction with verbs in the ます-form to express desire (**…たい**), with verbs in the て-form to express hope for someone else (**…てほしい**), or with nouns to express a desire for possession (**N がほしい**).

### 2. **Conjugation/Transformation Rules:**

- **V[ます-form]たい ものだ**: Used to express the speaker's desire to do something.
