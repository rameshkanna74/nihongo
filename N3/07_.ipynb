# Grammar 1


1.V conditional form よかった。/ V[ない-form]なければよかった。

I wish I had done … / I shouldn't have done …


“V conditional form よかった” reveals a regret of the speaker, feeling that it could
come to a better result if the speaker had done a certain thing.

“のに” can be added to the end.

“V conditional form よかったのに” and “V[ない-form]なければよかったのに”.

1. 友 達 は 私 の 話 に 傷 ついたみたい。あんなことを言わなきゃよかった。
   My words seem to have hurt my friend. I shouldn't have said that to him.

2. 昨日の 学 園 祭 に来ればよかったのに。とても 楽 しかったよ。
   You should have come to our school festival. It was really fun.

3. あんなに 恐 ろしい映画、見なければよかったなあ。
   I shouldn't have watched such a scary movie.

The sentence patterns below can also represent a feeling of satisfaction or
no regrets about what a speaker has done.

V[て-form] よかった
V[ない-form]なくて よかった

午後になって、 雨 が降り出した。 傘 を持ってきてよかった。
It started raining in the afternoon. I am glad that I brought an umbrella.

けがをしなくてよかった。
I am glad that I wasn't injured.


### 1. **Explanation:**

The grammar points **V conditional form よかった** and **V[ない-form]なければよかった** express regret or satisfaction regarding past actions. The former indicates a wish that something had been done differently, while the latter conveys relief that something was not done. These expressions are used to reflect on decisions or events that occurred in the past, allowing speakers to express their feelings about those experiences.

- **V conditional form よかった**: "I wish I had done ..."
- **V[ない-form]なければよかった**: "I shouldn't have done ..."

Both can also include **のに** to emphasize a sense of regret or dissatisfaction.

### 2. **Conjugation/Transformation Rules:**

- **For V conditional form (よかった)**:

  - Take the **plain form** of the verb (e.g., **行く** for "to go") and attach **よかった**.
    - Example: 行く → 行けばよかった (I wish I had gone).

- **For V[ない-form]なければよかった**:
  - Convert the verb into its negative form and attach **なければよかった**.
    - Example: 行かない → 行かなければよかった (I shouldn't have gone).

# Grammar 2


2.…わけだ
Of course/expected result

① “わけだ” means that “it is natural…” and “it is a matter of course…”.

It is used to express a natural consequence based on facts and is often used with “…ので/だから”.

V plain form わけだ

い-adj わけだ

な-adj な わけだ

N な/である わけだ

1. 夫 は仕事をしているが、子供が 4 人 もいるので、 私 も 毎 日 頑張って 働 いているわけです。
   Because we have four kids, even though my husband works, as you'd expect, I work really hard every day.

2. 彼 はアメリカで 10 年 も 働 いていたから、英語が 上 手 なわけです。
   He has been working in the USA for 10 years, so of course he can speak good English.

3. 消 費 税 があがったので、物価が 高 いわけだ。
   Because the consumption tax is high, the price is high.

4. 山田さんは 毎 月 ブランドのカバンを買っている。金持ちなわけだ。
   Mr. Yamada buys branded bags every month. Certainly he is a rich man.


### 1. **Explanation:**

The grammar point **…わけだ** is used to express a natural conclusion or an expected result based on preceding facts. It implies that the situation or outcome being discussed is logical or natural given the circumstances. It is often used in conjunction with phrases like **…ので** or **…だから** to clarify the cause of the conclusion.

### 2. **Conjugation/Transformation Rules:**

- **V plain form + わけだ**: Used for verbs.
- **い-adj + わけだ**: Used for i-adjectives.
- **な-adj + な + わけだ**: Used for na-adjectives, attaching な before わけだ.
- **N + な/である + わけだ**: Used for nouns, attaching な or using である for formal contexts.

3.…わけがない/わけはない

# Grammar 4


4.…させられる
Causative Passive form

A causative-passive form is used to express causative sentences with a passive voice.

In other words, it is the passive form of causative verbs.

This is used when a subject is a causee who is forced to do something.

More than a causative sentence, this implies the causee's spontaneous emotion and feelings of being victimized or forced.

1.昨日 野 球 の 練 習 に行かなかったら、今日コーチに グラウンドを 何 周 も走 らされた。
Because I didn't go to baseball practice yesterday, my coach made me run many laps around the field today.

2.ピーマンが 大 嫌 いだけど、 母 に食べさせられた。
I really hate green peppers, but my mom made me eat them.

3.彼氏と午前9時に 約 束 したのですが、駅 で 2時間も待たされました。本 当 に 腹 が立ってしょうがない。
I made a promise with my boyfriend to meet at 9:00 a.m., but he made me wait 2 hours for him at the station.
I cannot help but to feel angry.

### 1. **Explanation:**
The causative passive form, represented as "させられる" (saserareru), combines both causative and passive constructions in Japanese. This form is used when the subject (the causee) is forced to perform an action by someone else, indicating a sense of victimization or lack of choice. It highlights the emotional aspect of being compelled to do something against one’s will, making it distinct from simple causative forms. This construction is significant as it conveys both the action and the emotional impact of being coerced into performing that action.

### 2. **Conjugation/Transformation Rules:**
To form the causative passive, follow these steps based on the verb type:

1. **Group 1 Verbs (u-verbs)**: Change the final "u" to "aseru" (causative form) and then to "saseru" (passive form).
   - Example: 飲む (nomu) → 飲ませる (nomaseru) → 飲まされる (nomasareru)

2. **Group 2 Verbs (ru-verbs)**: Drop the "ru," change to "saseru."
   - Example: 食べる (taberu) → 食べさせる (tabesaseru) → 食べさせられる (tabesaserareru)

3. **Irregular Verbs**: 
   - する (suru) → させる (saseru) → させられる (saserareru)

# Grammar 5


5.…こそ

Just/Very … (emphasizes the preceding word)

“こそ” is used to emphasize a certain thing that is derived from many things.

It follows nouns and particles.

“からこそ” is used to emphasize the cause and reason.

1. 今 回 の 実 験 の 成 功 は 運 じゃない。 努 力 したからこそ、 成 功 したんだ。
   The success of this experiment is not because of luck.
   It was attained as the result of efforts.

2. 今年こそ試験に 合 格 できるよう、頑張ります。
   I hope this will be the year that I can pass the exam. I will try my best.

3. この試合は、みなさんの 応 援 があったからこそ勝つことができました。
   This game could be won based solely on everyone's support.


### 1. **Explanation:**

The particle "こそ" (koso) is used in Japanese to emphasize a particular word or phrase, highlighting its importance or significance. It is often translated as "just," "only," or "precisely." This emphasis can apply to nouns or phrases that are preceded by certain particles (like "が," "は," "も"). When used in conjunction with "からこそ" (kara koso), it emphasizes the reason or cause for a situation. This grammatical structure plays a significant role in conveying the speaker's intent and emotional nuance.

### 2. **Conjugation/Transformation Rules:**

"こそ" is not conjugated in the traditional sense, but it must be placed directly after the word it emphasizes. Here are some common usages:

- **Nouns**: Noun + こそ
  - Example: 友達こそ (tomodachi koso) - "it is friends (that...)"
- **Particles**: Particle + こそ
  - Example: からこそ (kara koso) - "because (it is)..."

6.N から言うと/から言えば/から言 って
From someone's point of view

This is used when mentioning a thought, a conclusion or a decision made by someone.

1. この 成 績 から言えば、国 立 大 学 は 厳 しいと 思 います。もっと頑張りなさい。
   Judging by your score, it will be hard to enter a national university.
   Please study harder.

2. 今 月 の 注 文 状 況 から言って、 売 上 はかなり 高 くなるだろう。
   With this monthʼs product orders, I think sales should be very high.

   From the customer's viewpoint, there is a small number of product

### 1. **Explanation:**

The expression "N から言うと" (N kara iu to), "から言えば" (kara ieba), or "から言って" (kara itte) is used to indicate a perspective or viewpoint based on a certain criterion or context. It allows the speaker to express thoughts, conclusions, or evaluations from a specific point of view. This construction is significant in Japanese as it acknowledges different perspectives and provides a framework for discussing opinions or assessments based on particular conditions.

### 2. **Conjugation/Transformation Rules:**

The structure generally follows the pattern:

- **N から言うと**: Used for stating a viewpoint based on a noun (N).
- **N から言えば**: Similar usage, slightly more formal or hypothetical.
- **N から言って**: More conversational and used in everyday speech.

**N** can be a noun, typically indicating a person, role, or perspective (e.g., 客 (kyaku) - customer, 学生 (gakusei) - student).

# Grammar 7


7.…たって/だって

“…たって/だって” are the spoken words of “…ても/でも”.

There are two ways of using this grammar.
① “…たって/だって” exemplifies a certain thing, and indicates that it is

### 1. **Explanation:**
The grammar point “…たって/だって” is a colloquial expression in Japanese that conveys the meanings of "even if," "no matter how," or "everyone." It serves as a spoken alternative to the more formal “…ても/でも.” This construction is useful for expressing hypotheticals or general statements that emphasize equality or universality.

### 2. **Conjugation/Transformation Rules:**
To use “…たって/だって,” follow these transformation rules based on the type of word you're starting with:

- **Verb (V) [た-form]:** Take the verb and convert it into the past form (た-form).  
  Example: 食べる (taberu) → 食べた (tabeta)
  
- **I-adjective (い-adj):** Remove the final い and add たって.  
  Example: おいしい (oishii) → おいしくたって (oishikuttatte)
  
- **Na-adjective (な-adj):** Attach だって after the adjective.  
  Example: 静かだ (shizukada) → 静かだって (shizukadatte)
  
- **Noun (N):** Attach だって after the noun.  
  Example: 先生 (sensei) → 先生だって (senseidatte)

# Grammar 10

Every time …

# Grammar 11


11.…に 比 べ(て)/と 比 べ(て)

Compared to/with…

This grammar is used to compare things that are related to each other.

Either particles “に” or “と” can be used.

1. 去 年 に 比 べて、 身 長 が 5 センチも伸びました。
   Compared to last year, I am 5 centimeters taller.

1. 地 球 は 太 陽 に 比 べて 小 さい。
   The earth is small in comparison with the sun.

1. この 町 で石油が 発 見 されてから、 経 済 の 発 展 はとても 速 かったです。5 年 前 と 比 べてずいぶんにぎやかになりました。

   It has become quite livelier than it was five years ago.


### 1. **Explanation:**

The grammar point **…に比べて** (ni kurabete) or **…と比べて** (to kurabete) is used to express comparison between two or more entities. It highlights how one item is different from or similar to another in some aspect. This construction is significant in Japanese as it allows for nuanced discussions about similarities and differences in various contexts, such as size, quantity, quality, or performance.

### 2. **Conjugation/Transformation Rules:**

- For nouns:
  - **Noun** + に比べて (ni kurabete) / と比べて (to kurabete).
- For verbs or adjectives, the construction remains the same, using the relevant noun form as a reference point:
  - Example: **V** (plain form) + に比べて / と比べて.
