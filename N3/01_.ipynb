N3 Lesson 1 grammar

1. 縮 約 形
   Contractions
   Contractions are the omission of sounds and the changing of the proper grammar
   form into casual conversation.
   “じゃ” and “やってた” are such examples of contractions of casual spoken words
   used in the Japanese language.
   The following are a list of six basic rules on how to properly conjugate and change
   words to form proper contractions.

①The changing rule of “は”:
では → じゃ
ては → ちゃ

Examples: 学 生 ではありません → 学 生 じゃありません

見てはいけない → 見ちゃいけない

②The changing rules of “て” and“で”:
ている → てる
でいる → でる
ておく → とく
でおく → どく
てしまう → ちゃう
しまう → じゃう

Examples:
読んでいる → 読んでる
置いておく → 置いとく
忘 れてしまった → 忘 れちゃった
飲んでしまった → 飲んじゃった

③The changing rule of the conditional form “ば”:
すれば → すりゃ

Examples: 昨日 掃除すればよかった → 昨日掃除すりゃよかった

④The changing rule of the citation “と”:(Notice the double consonant)
と → って
という → っていう

Examples: この 花 はバラと言います → この 花 はバラって言います

- **Obligation or Necessity:** This grammar point is used when there is a duty or requirement to do something.
- **Contracted Form:** "なきゃ" is a contracted form of "なければ" and is more commonly used in casual speech.
- **Flexibility:** While the full form "なければなりません" is more formal, the contracted form "なきゃ" is versatile and can be used in various situations.
- **Emphasis:** This grammar point can be used to emphasize the importance or urgency of a task.

### Examples and Explanations
