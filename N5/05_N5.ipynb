### 1. **Explanation:**

In Japanese, verbs are divided into three main groups: **Group 1**, **Group 2**, and **Group 3**. Understanding these groups is important because it affects how verbs are conjugated into different forms, such as past tense, negative, and polite forms. The classification is based on how the verb ends in its dictionary (plain) form and helps learners predict how to modify verbs for different grammatical situations.

- **Group 1 (五段動詞 - Godan-doushi)**: These verbs end in **"u"** sounds like "ru," "su," "mu," etc. Their stem changes when conjugated.
- **Group 2 (一段動詞 - Ichidan-doushi)**: These verbs mostly end in **"-iru"** or **"-eru"**, and their stem remains consistent across conjugations.
- **Group 3 (不規則動詞 - FUKISETSU-doushi)**: There are only two verbs in this group, **くる (to come)** and **する (to do)**, along with all verbs ending in **する**.

### 1. **Explanation:**

The **ます-form** (polite form or ていねいけい) is one of the most important verb forms in Japanese. It is used to convey politeness and respect, making it appropriate for use in formal settings, such as conversations with strangers, superiors, or in work environments. The **ます-form** does not change based on whether the action is directed at the speaker or listener—it is simply used to be polite.

In Japanese, verbs conjugated in the **ます-form** have two main tenses:

- **Non-past tense**: This form can indicate both **present** and **future** actions.
- **Past tense**: This form expresses actions that have already occurred.

- **ません**: Negative non-past (present/future negative)

### 2. **Conjugation/Transformation Rules:**

The **ます-form** can be derived from the dictionary form of verbs. Below is how to conjugate verbs from the dictionary form to their corresponding **ます-form** for each group:

1. Change the final **"-u"** sound of the verb to the corresponding **"-i"** sound.
2. Add **ます** to the transformed verb.

- **のむ (nomu)** → **のみます (nomimasu)** (to drink)

#### **Group 2 (Ichidan Verbs)**:

1. Remove the final **"-ru"** from the dictionary form.
2. Add **ます** to the verb stem.

Example:

- **たべる (taberu)** → **たべます (tabemasu)** (to eat)

#### **Group 3 (Irregular Verbs)**:

1. **する** becomes **します** (to do).
2. **くる** becomes **きます** (to come).
