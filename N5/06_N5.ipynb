# Grammar 7


7．N (Place) に  V

…to somewhere

The particle “ に ” indicates the destination. “Place + に + verb”is the sentence pattern. 

“に” follows a place name and is followed by a movement verb.

1. きんようび かいしゃに いきます。
I will go to the company on Friday.

2. きょねん インドに いきました。
I went to India last year.

3. ラジさんは 4 がつから だいがくに いきます。
Raj will go to a university in April.

Sure! Here’s a detailed explanation of the grammar point **N (Place) に V**:

### 1. **Explanation:**

The particle **に** indicates the destination or target of an action. When used in the structure **N (Place) に V**, it specifies where an action is directed or where someone or something is moving towards. This construction is essential for expressing movement or intent to go to a particular place.

### 2. **Conjugation/Transformation Rules:**

- The verb following **に** must indicate movement, such as:
  - **行く (いく, iku)** - to go
  - **来る (くる, kuru)** - to come
  - **帰る (かえる, kaeru)** - to return
- The place can be any noun indicating a location (e.g., school, home, park).
