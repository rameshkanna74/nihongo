# Grammar 1


1.V1 [た-form］り、V2 [た-form］りします。

Giving two and two more examples of many actions.

In this pattern, the verb’s た-form is used. 

This sentence pattern is used when you give two or more examples of various actions.

The tense of the sentence is expressed at the end.

1. A: 家 へ 帰 ってから、よく 何 を していますか。
   What do you often do after you come back home?

B:テレビを 見たり、 宿 題 を したり、漫画を 読んだりしています。
I watch TV, do the homework, and read the comic books and so on.

2. この部屋で タバコを 吸ったり、食べ 物 を 食べたり しないで ください。
   Please don't smoke and eat food or some things like that in this room.


## 1. Explanation:

The Japanese grammar pattern **V1 [た-form]り、V2 [た-form]りします** is used to describe various actions or activities by providing examples. This structure allows speakers to convey multiple actions in a concise manner while indicating that there are more examples that could be given. It emphasizes the variety of activities without needing to list them exhaustively. The tense of the sentence is determined by the final verb form.

## 2. Conjugation/Transformation Rules:

To use this structure, follow these steps:

1. **Identify the verb**: Start with the base form of the verb (the dictionary form).
2. **Convert to the た-form**:
   - For **group 1 (u-verbs)**: Change the final う sound to った, いった, etc. For example, 行く (iku) becomes 行った (itta).
   - For **group 2 (ru-verbs)**: Drop the る and add た. For example, 食べる (taberu) becomes 食べた (tabeta).
   - For **irregular verbs**:
     - する (suru) becomes した (shita).
     - 来る (kuru) becomes 来た (kita).
3. **Combine the た-forms**: Use たり between the た-forms of the verbs, and end with します or する for the final action.

### Example Transformations:

- 食べる (taberu) → 食べた (tabeta)
- 飲む (nomu) → 飲んだ (nonda)
- 読む (yomu) → 読んだ (yonda)

## 1. Explanation:

## 2. Conjugation/Transformation Rules:

To use this structure, follow these steps:

1. **Identify the verb**: Start with the base form of the verb (the dictionary form).

   - For **irregular verbs**:
     - する (suru) becomes します (shimasu).
     - 来る (kuru) becomes 来ます (kimasu).
3. **Combine the ます-form with ながら**: Use the ます-form of the first verb followed by ながら, and then add the second verb in its appropriate form (plain or polite).

### Example Transformations:

- 食べる (taberu) → 食べます (tabemasu)
- 飲む (nomu) → 飲みます (nomimasu)
- 読む (yomu) → 読みます (yomimasu)

N1 represents the subject. N2 represents the noun that expresses ability,

- The sentence can express past or ongoing situations based on the verb form used.
- Ensure to use the correct form of the verb or adjective to match the time context (past, present).
- Use **のとき** with nouns to specify the time related to the noun.
