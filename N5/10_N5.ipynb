# Grammar 1


1. P へ V [ます form] に 行きます/来ます/ 帰 ります。

   P へ N に 行きます/来ます/ 帰 ります。

These three patterns express purposes of going/coming/returning.

This represents the purpose of going somewhere for a certain objective.

The sentence pattern is as follows: Start with the place name, followed by the
particle “へ” (which indicates a direction), then put either a noun or the “ます
-form” of a verb, followed by the particle “に”, and end with a verb.

あきはばらへ まんがを 買いに 行きます。

日本へ 日本語の べんきょうに 行きます。

1. 本屋へ 本 を 買いに 行きます。
   I will go to the book store to buy a book.

2. アメリカへ 英語の 勉 強 に 来ました。
   I came to America to study English.

3. うちへ 母 の料 理 を 食べに 帰 ります。

I came back home to my mother's cooking.

Note: There're 2 rules when the verb is “noun + する” which belongs to Group 3.

①When the verb has no object: don’t use the ます-form, but use only the noun.

For example,

べんきょうする(to study)

“べんきょう” is a noun “べんきょうし” is the ます-form.

When the verb is “べんきょうする” by itself, don’t use “べんきょうし” but use the noun “べんきょう”.

② When inserting an object into Group 3 verbs, use the ます-form.

For example,

にほんご日本語をべんきょうする(to study Japanese)

Here, you use “日本語をべんきょうし”.


### 1. **Explanation:**

The grammatical structure "P へ V [ます form] に 行きます/来ます/帰ります" is used to express the purpose of going, coming, or returning to a specific place. This construction highlights the intention behind the movement, clarifying why the speaker is going to that location. It is significant in Japanese as it allows for clear communication of goals and motivations related to travel or movement.

### 2. **Conjugation/Transformation Rules:**

- **Basic Structure:**
  - Place name + へ (e.g., 本屋へ) + [noun or verb in ます-form] + に + 行きます/来ます/帰ります.
- **Noun as Purpose:**

  - If the verb is a noun (e.g., 勉強), you simply use the noun itself (勉強に行きます).

- **Verb as Purpose with Object:**

  - If the verb is a noun + する (e.g., 日本語を勉強する), use the ます-form (勉強します) when there is an object.

- **Verb as Purpose without Object:**
  - When using a noun + する without an object, just use the noun (e.g., 勉強に行きます).

9.V[ない form]なければなりません。
I have to do….
The meaning of “V なければなりません” is “have to do/ must do”. It means the
person must do something. Conjugation relies on the ない-form of the verb.

1.明日 朝 6時に 起きなければ なりません。
I have to get up at 6:00 tomorrow morning.

2.金 曜 日 までに DVD を 返 さなければ なりません。
I have to return this DVD by Friday.

Here's a detailed explanation of the grammar point "V[ない form]なければなりません":

### 1. **Explanation:**

The expression "V[ない form]なければなりません" is used to express obligation or necessity. It translates to "have to do" or "must do" in English. This construction indicates that the speaker is required to perform a certain action. It's important in both formal and casual contexts, allowing speakers to express duties, rules, or responsibilities.

### 2. **Conjugation/Transformation Rules:**

To use this construction, follow these steps:

1. Start with the verb in its dictionary form (e.g., 行く, 食べる).
2. Convert it to its ない-form (e.g., 行かない, 食べない).
3. Attach "ければなりません" to the ない-form.

**Examples of conjugation:**

- 行く (iku) → 行かない (ikanai) → 行かなければなりません (ikanakereba narimasen) - I have to go.
- 食べる (taberu) → 食べない (tabenai) → 食べなければなりません (tabenakereba narimasen) - I have to eat.

# Grammar 10

10.助詞の 省 略

Omission of particles

In Japanese, post-positional particles are often omitted on casual occasions.

Especially if the speaker and listener are close, it sounds more natural to
omit post-positional particles “が”, “は”, “を”, and “に”.

For example, この映画がは とても おもしろいですね。

During a conversation among friends, “は” can be omitted.

この映画、 とても おもしろいですね。

いっしょに ごはんを たべますか。

Omit “を”, then it becomes:

いっしょに ごはん たべますか。

This is frequently used on casual occasions.

### 1. **Explanation:**

In Japanese, the omission of particles (助詞の省略) occurs frequently in casual conversation, particularly among close friends or family members. This practice makes the speech sound more natural and fluid. Commonly omitted particles include "が," "は," "を," and "に." While omitting particles can lead to a more informal tone, it is important to maintain clarity in communication. This grammatical feature reflects the contextual understanding shared between speakers.

### 2. **Conjugation/Transformation Rules:**

While there aren't specific transformation rules for omitting particles, the following considerations are important:

- When omitting particles, ensure the remaining sentence still conveys the intended meaning.
- The context should be clear enough for the listener to understand without the omitted particles.
