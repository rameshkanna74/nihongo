# Grammar 1


1.もう V［ます-form］ ました。

Someone has already done something.

“もう” shows that something has been completed, so it means “already” in English.

The verb that follows “もう” is conjugated into the past tense.

The interrogative form: もう V［ます-form］ましたか。

Have you already done something?

The positive response:

はい、もう V［ます-form］ました。

Yes, I've already done….

The negative response:

いいえ、V[て-form]いません。 or  いいえ、まだです。
No, I haven't done …yet. Not yet.

1. A:山田さん、もう 本 を 買いに 行きましたか。
Mr. Yamada, have you gone out and bought the book yet?

B:いいえ、まだ 買っていません。
No, I haven't bought it yet.

2. A:もう 手紙を 送 りましたか。
Have you sent the letter?

B:はい、もう 送 りました。
Yes, I have sent it.

Here’s a detailed explanation of the grammar point "もう V［ます-form］ました":

### 1. **Explanation:**

The phrase "もう V［ます-form］ました" indicates that an action has already been completed. "もう" means "already," emphasizing that the speaker is confirming the completion of an action. This construction is frequently used in conversational Japanese, especially when discussing tasks, errands, or actions that have a specific timeframe.

### 2. **Conjugation/Transformation Rules:**

To use this grammar point, follow these steps:

- Take the verb in its ます-form (the polite form).
- Change it to the past tense by replacing the ます with ました.

**Examples of Transformation:**

- 食べます (tabemasu - to eat) → 食べました (tabemashita - ate)

# Grammar 2


2.目 的 語 の 強 調

Making an object the topic of a sentence
When stressing on an object word, the word order changes. Usually, it is
“object + を + verb”. When the object word becomes the sentence topic, the
object is inserted at the beginning of the sentence to show emphasis. “を”
becomes “は”.

The order is “object word + は + verb”.



### 1. **Explanation:**

In Japanese, when you want to emphasize the object of a sentence, you can change the structure from the standard "object + を + verb" to "object + は + verb." This shift in word order makes the object the topic of the sentence, thereby highlighting its importance. This construction is particularly useful in conversations where you want to clarify or stress the object being discussed.

### 2. **Conjugation/Transformation Rules:**

To transform a sentence where the object is the focus:

- Identify the object in the sentence.
- Replace the particle "を" with "は."
- Move the object to the beginning of the sentence.

**Example:**

- Standard: りんごを 食べます (Ringo o tabemasu) - "I eat an apple."
- Emphasized: りんごは 食べます (Ringo wa tabemasu) - "As for the apple, I eat (it)."

# Grammar 8


8. … 前 に V

This is also used when describing the order of actions. Before “ 前 ”, insert the

V dictionary form 前 に V

Certainly! Here’s a detailed explanation of the grammar point "…前に V" (Before doing…, do…).

### 1. **Explanation:**

The structure "V dictionary form 前に V" or "N の前に V" is used to describe actions that occur before another action takes place. It emphasizes the order of actions, indicating that one event needs to be completed before another begins. This structure is significant in Japanese as it helps clarify timelines in conversation and writing.

### 2. **Conjugation/Transformation Rules:**

- **For Verbs (V):** Use the dictionary form of the verb before "前に."
  - Example: 食べる (taberu - to eat) → 食べる前に (taberu mae ni - before eating)
- **For Nouns (N):** Add "の" before "前に."
  - Example: 旅行 (ryokou - trip) → 旅行の前に (ryokou no mae ni - before the trip)

# Grammar 9


9. V [た-form] 後で …

After doing…, do…

This grammar is also used when indicating an action's order.

V た-form 後 でV
N の後 でV

1. 空 港 で チェックインした 後 で、買 い 物 をします。
I will go shopping after I check in at the airport.

2.仕事が 終わった 後 で、 友 達 と お酒 を 飲みに 行きました。
I went to drink alcohol with my friend after work.

3. 授 業 の 後 で 休 みます。
I rest after class.

### 1. **Explanation:**

The construction "V [た-form] 後で…" is used to indicate that one action occurs after another. It is similar to the English phrase "after doing...". This structure emphasizes the sequence of actions and is commonly used in everyday conversation.

### 2. **Conjugation/Transformation Rules:**

- To create the [た-form] of a verb, you typically change the verb into its past tense form.

  - For example:
    - 行く (iku - to go) → 行った (itta)
    - 食べる (taberu - to eat) → 食べた (tabeta)
    - 飲む (nomu - to drink) → 飲んだ (nonda)

- For nouns, the structure is "N の後で V":
  - Example: 学校の後で勉強する (gakkou no ato de benkyou suru - to study after school).
