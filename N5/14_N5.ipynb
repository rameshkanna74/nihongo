# Grammar 1


1.V[て-form] ください。

Please do….

This grammar is used when you request or recommend someone to do something.

In English, it means “please do something”.

However, you don't say it to your manager or boss, because it sometimes sounds like an order or instruction.

1. この部屋は とても 暗 いです。 電気を つけて ください。
   This room is very dark. Please turn on the light.

2. 2 年 生 は 手を あげて ください。
   Sophomores, please raise your hand.

3. 家族の 写 真 を 見せて ください。
   Please show me your family picture.

4. この単語を 覚 えて ください。
   Please remember this word.


Sure! Here's a detailed explanation of the grammar point **V[て-form] ください**.

### 1. **Explanation:**

The phrase **V[て-form] ください** is used to make polite requests or recommendations in Japanese. It translates to "Please do [verb]" in English. This structure is common in everyday conversation when asking someone to perform an action, making it an essential part of polite communication. However, it's important to note that it can sound commanding, so it is typically not used with superiors or in formal business settings.

### 2. **Conjugation/Transformation Rules:**

To form the て-form of a verb:

- **Group 1 (U-verbs)**: Change the final -u sound to -tte. (e.g., かく (kaku) → かいて (kaite))

# Grammar 9


9.きかんに～ 回

～times in one period

Before the particle “に”, put a noun expressing the period, and it is “Number+かい”. 

It represents frequency.

1. 1 年 に 2 回 旅 行 します。
I travel twice a year.

2. 3 分 に 1 回 電 車 が 来ます。
The train comes once every three minutes.

3. 毎 週 1 回 東 京 に 行きます。
I go to Tokyo once a week.

### 9. きかんに～回

#### 1. **Explanation:**

The construction "きかんに～回" (kikan ni ~kai) is used to express frequency within a specified period. This structure allows speakers to indicate how many times an action occurs within a given timeframe, making it essential for discussing routines, schedules, and habitual actions. It's significant in daily conversations, particularly when discussing travel, meetings, or any regular activity.

#### 2. **Conjugation/Transformation Rules:**

To use this structure:

1. Choose a period noun (e.g., 年 "year," 月 "month," 週 "week," 日 "day").
2. Place it before the particle "に."
3. Use the number followed by "回" (kai) to indicate frequency.

Example:

- 1 回 (ikkai) = once
- 2 回 (nikai) = twice
- 3 回 (sankai) = three times
