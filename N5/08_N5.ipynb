4.い-adj N です。
It is ... N.

Here, the “い-adjective” modifies a noun.

The sentence structure is “い-adjective N”.

Insert the “い-adjective” before the noun. When there’s a subject in the sentence, the sentence pattern is “N1 is い-adjective N2”.

1. ジャパンホテルは あの 白 い 建 物 です。
   Japan Hotel is that white building.

2. この 紅 茶 は とても おいしいお 茶 です。
   This black tea is really delicious tea.

3. 今日は いい天気です。
   It's good weather today.


Certainly! Let’s delve into the structure “い-adj N です” in Japanese.

### 1. **Explanation:**

The structure “い-adj N です” is used to describe a noun using an “い-adjective.” In this construction, the adjective directly modifies the noun that follows it. This pattern is crucial for forming descriptive sentences in Japanese, allowing speakers to convey characteristics of nouns succinctly. The addition of “です” at the end provides politeness and formality to the statement.

### 2. **Conjugation/Transformation Rules:**

While the adjective itself doesn’t change form in this construction, it’s essential to know how to use “い-adjectives” in different contexts:

- **Non-Past Affirmative:** い-adjective (e.g., 白い)
- **Non-Past Negative:** い-adjective くない (e.g., 白くない)
- **Past Affirmative:** い-adjective かった (e.g., 白かった)
- **Past Negative:** い-adjective くなかった (e.g., 白くなかった)

However, in the structure “い-adj N です,” the adjective remains in its dictionary form.

# Grammar 5


5.N は どうですか。

How is...?/ How do you think...?

This sentence pattern is used when asking the listener’s opinions, ideas, or impressions. 

When answering the question “どうですか”, an adjective is included in the response.

1N は どうですか。

How is...?/ How do you think...?

1. A:お仕事は どうですか。
How is the job?

B:いそがしいです。
It is busy.

2. A:このコーヒーは どうですか。
What do you think of this coffee?

B:とても おいしいです。
It is very delicious.


2N は どうでしたか
How was...?
When you ask the listener’s opinions, ideas, and impressions about a past
incident, you ask with “どうでしたか”. This is the past tense of “どうですか”.

1. A:昨日の 仕事は どうでしたか。
How was work yesterday?

B:あまり 忙 しくなかったです。
It wasn't so busy.

2.  A:旅行はどうでしたか。
How was the trip?

B:楽しかったです。
It was fun.

Here’s a detailed explanation of the grammar point "N は どうですか" and its past form "N は どうでしたか."

### 1. **Explanation:**

The phrases “N は どうですか” and “N は どうでしたか” are used to inquire about the listener's opinions or impressions regarding a noun (N). The first structure asks about the current state or condition, while the second asks about a past state. These phrases are essential in conversational Japanese, as they invite feedback and encourage discussion.

### 2. **Conjugation/Transformation Rules:**

- **Present Form:**
  - N は どうですか
- **Past Form:**
  - N は どうでしたか

In responses, adjectives are typically used to express opinions or feelings, and these adjectives can be in the non-past or past forms as necessary.

# Grammar 6


6.N は い-adj (...い) くて
It is ..., and ....
When you consecutively say two or more adjectives, you use the て-form. The
proper conjugation is “い-adjective (...い) くて...”. When using two adjectives
consecutively to describe past events, only the adjective at the end of the
sentence is changed to the past tense.

1.この部屋は 新 しくて 広 いです。
The room is new and wide.

2. 昼 ごはんは 安 くて おいしいです。
The lunch is cheap and delicious.

3. A:きのうの 天気は どうでしたか。
How was the weather yesterday?

B:天気が わるくて、とても寒 かったです
It was really bad and cold.


Here’s a detailed explanation of the grammar point "N は い-adj (...い) くて" used for connecting two adjectives.

### 1. **Explanation:**

The structure "N は い-adj (...い) くて" is used to connect two or more adjectives to describe a noun (N). It allows for a more nuanced description by linking qualities together. This construction is important for providing detailed information and can make speech and writing more fluid.

### 2. **Conjugation/Transformation Rules:**

- To connect two **い-adjectives** using this structure, take the **-い** form of the adjective and change it to the **-くて** form.
- The adjective that appears last in the sentence can be conjugated for tense (present or past), while the preceding adjectives remain in their original forms.

**Examples of Conjugation:**

- 新しい (new) → 新しくて (new and)
- 高い (expensive) → 高くて (expensive and)
- 寒い (cold) → 寒くて (cold and)

# Grammar 7


7.N は い-adj (...い)く V

Using い-adjectives to describe verbs

This is how to properly conjugate an い-adjective into an adverb.

When an adjective modifies a verb, it has to be changed into the adverbial usage form. 

The sentence pattern is “N is い-adjective (...い) く V”. 

The conjugation rule is to replace “い” of “い-adjective” words with “く”. 

Then, add a verb after the conjugated word. “い-adjective (...い)く V”

1.昨日 早 く 寝ました。
I slept early yesterday.

2. 電 車 は 遅 く 来ました。
The train came late.

3. 友 達 と 楽 しく 旅 行 しました。
I went on a fun trip with my friend.

Sure! Here's a detailed explanation of the grammar point "N は い-adj (...い)く V":

### 1. **Explanation:**

The structure "N は い-adj (...い)く V" allows you to use い-adjectives to modify verbs. In Japanese, adjectives can be converted into adverbs to describe how an action is performed. This construction emphasizes the manner in which the action takes place, providing context to the verb by linking it to the quality expressed by the adjective.

### 2. **Conjugation/Transformation Rules:**

To transform an い-adjective into its adverbial form, follow these steps:

- Remove the final い (い).
- Add く to the stem.
- Place the resulting adverb before a verb.

**Example:**

- 高い (takai - high) → 高く (takaku - highly)
- 速い (hayai - fast) → 速く (hayaku - quickly)

### 3. **Examples and Explanation:**

# Grammar 8


8....が、...

...,but...

Here, “が” is a conjunctive particle used to connect ideas or express a contradiction. 

In English, it means “but”. It is usually used as “...ですが、...” or “... ますが、...”.

If there is an adjective inserted in a sentence that precedes or follows “が”,
the preceding/following sentence of “が” is contradictory; one is positive and
the other one is negative.

1 adjective sentences

1.漢字の勉強は 難しいですが、楽しいです。
Learning kanji is difficult, but fun.

2.フランスのお酒は おいしいですが、高いです。
French alcohol is delicious, but expensive.

3.わたしの家は 安いですが、狭いです。
My home is cheap, but small.

2verb sentences

I will go to Kyoto tomorrow, but I will come back to Tokyo at night.

2. 家 から 駅 までは 遠 いですが、バスが あります。
Even though my home is far from the station, there is a bus.

3. 学 校 に 行きましたが、 先 生 に 会いませんでした。
I went to school, but I did not meet the teacher.

### 8. ...が、...

#### 1. **Explanation:**

The particle “が” is a conjunction used in Japanese to connect two ideas or clauses, often indicating contrast or contradiction. It translates to “but” in English. This structure allows speakers to present a positive statement followed by a negative one, or vice versa, thus highlighting differing aspects of a situation.

#### 2. **Conjugation/Transformation Rules:**

- **Adjectives:** For い-adjectives, the sentence preceding “が” should typically use the plain form (non-past), while the following statement can also be in plain form (negative).
  - Example: _難しいですが_ (It is difficult, but...)
- **Verbs:** For verbs, the first clause should be in the plain form (non-past), and the second clause follows similarly.
  - Example: _行きますが_ (I will go, but...)

# Grammar 9


9. N が ...。
   N is .... / N does....

Here, “が” is the subject of the predicative. In Lesson 1, you studied “は”. 

In this lesson, “が” is almost identical to “は” when indicating the subject. 

The difference is that the “は” is the primary particle used to indicate a subject.

When “が” indicates the subject, it has a special meaning.

1. N1 が N2 です。
   N1 is N2.

   N2 is being searched for, and describes N1. 1.ジャパンホテルは どれですか。
   Which one is Japan Hotel?


   あの白い建物が ジャパンホテルです。
   That white building is Japan Hotel.

2. 山田さんは 誰 ですか。
   Who is Mr. Yamada.

あの人が 山田さんです。
He is Yamada.

2 N が V。

N does....

When putting stress on the subject N, “が” is used.

1. 金 さんが 大 阪 に 行きます。
   It is Mr. Kim who will go to Osaka.

2. 山田さんが コーヒーを 飲みました。
It is Mr. Yamada who drank coffee.


### 9. N が ...

#### 1. **Explanation:**

The particle “が” is used in Japanese to indicate the subject of a sentence. While similar to “は,” “が” emphasizes the subject or identifies it in a particular context. This construction can be used to describe or assert something about the subject.

- **N が N2 です**: This structure indicates that N1 is identified as N2, often when N2 is new information or something being searched for.
- **N が V**: This structure emphasizes the subject N and its action, highlighting who is performing the action.

#### 2. **Conjugation/Transformation Rules:**

There are no specific conjugation rules for using "が" as it serves primarily as a subject marker. The following general rules apply:

- The noun preceding “が” is the subject of the sentence.
- The verb or predicate that follows describes what the subject is doing or what it is identified as.
