### 2. **Conjugation/Transformation Rules:**

- **With Nouns:**
  - **Structure:** Noun + は + いつですか？
  - **Example:** **すずきさんの たんじょうびは いつですか？** (When is Mr. Suzuki's birthday?)

- **With Verbs:**
  - **Structure:** いつ + Verb
  - **Example:** **ラジさんは いつ きましたか？** (When did Raj come?)

### 4. **Important Points:**

- **Flexibility:** "いつ" can be used with both nouns and verbs to ask about when something happens.
- **Context:** Ensure that the context of the question is clear. For example, if asking about a future event, "いつ" should be used with a verb indicating that future action.
- **Formality:** The structure of the question remains the same regardless of the formality of the situation. However, politeness can be indicated by using more formal verbs and phrasing in responses.
